```python
import pandas as pd
import sqlalchemy as db
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

## Crearemos la clase ConnectionMySQL

```python
# este es el fichero para conectarnos a MySQL
class ConexionMySQL():
    def __init__(self, bd_servidor,bd_nombre,bd_usuario,bd_contrasenia):
        self.bd_servidor=bd_servidor
        self.bd_nombre=bd_nombre
        self.bd_usuario=bd_usuario
        self.bd_contrasenia=bd_contrasenia

    ## Función de conexión a MySQL
    def conectarMySQL(self):
        engine = db.create_engine(f"mysql+mysqlconnector://{self.bd_usuario}:{self.bd_contrasenia}@{self.bd_servidor}:3310/{self.bd_nombre}")
        conn=engine.connect()
        return conn
```

## Conexion MySql a nuestra bd olap o warehouse

```python
#ConexionMySQL
## creando el objeto de la clase ConnetionMysql
dw_systrxbanco = ConexionMySQL(
                            "172.27.224.1",            ## Servidor
                            "dw_systrxbanco",   ## base de datos
                            "root",                 ## usuario
                            "root"                 ## contraseña
                            )

## llamar a la funcion de conexión a la base de datos
conn_dw_systrxbanco = dw_systrxbanco.conectarMySQL()
```

## Creamos los dataframes de cada dimensión de la bd_warehouse

```python
df_fact = pd.read_sql("select * from FACT_TRANSACCIONES", con=conn_dw_systrxbanco)
df_fact["count"] = 1
df_dimcargoabono = pd.read_sql("select * from DIM_TIPCARGOABONO", con=conn_dw_systrxbanco)
df_dimsucursal = pd.read_sql("select * from DIM_SUCURSAL", con=conn_dw_systrxbanco)
df_dimcanal = pd.read_sql("select * from DIM_CANAL", con=conn_dw_systrxbanco)
df_tiempo = pd.read_sql("select * from DIM_TIEMPO", con=conn_dw_systrxbanco)
df_titran = pd.read_sql("select * from DIM_TIPOTRANSACCION", con=conn_dw_systrxbanco)
```

### 1.¿Cuántas transacciones se realizan mensualmente y diariamente durante el año 2023 ?

Transacciones diariametne

```python
#Transacciones diarimaente
transac_sem=df_fact[["dia_semana",'count']].groupby("dia_semana").count()
transac_sem.sort_values(by=["count"], ascending = False)
```

Transacciones Mensualmente

```python
#Transacciones mensualmente
transac_mensual=df_fact[["mes",'count']].groupby("mes").count()
transac_mensual=transac_mensual.sort_values(by=["mes"], ascending = True).reset_index()
transac_mensual
```

```python
#listmos solo los meses unicos con sus valores y nombre
df_tiempo_mesunico=df_tiempo.drop_duplicates(subset=['mes','mes_nombre'])
columns0=['mes','mes_nombre']
df_tiempo_mesunico=df_tiempo_mesunico[columns0]
df_tiempo_mesunico
```

```python
#Hacer merge para traer el nombre del mes
resul_transac_mensual=pd.merge(transac_mensual,df_tiempo_mesunico, on='mes',how='left')
#resul_transac_mensual2 = resul_transac_mensual[resul_transac_mensual['_merge'] == 'left']
resul_transac_mensual
```

```python
#visualizaciòn

import matplotlib.pyplot as plt

# Datos
x = resul_transac_mensual['mes_nombre']
y = resul_transac_mensual['count']

# Crear el gráfico de línea

plt.figure(figsize=(20,10))
plt.plot(x, y)
# Añadir etiquetas y título
plt.xlabel('Meses')
plt.ylabel('Cnt Trx')
plt.title('Tendencia de la cnt trx en los meses del año')

# Mostrar el gráfico
plt.show()
```

### 2.¿Cuántas transacciones de  abono se realizan por mes ?

```python
#creamos el dataframe de las trx
merge_fact_cargoabono=pd.merge(df_fact,df_dimcargoabono,left_on='DIM_TIPCARGOABONO_idcargoabonoSK',right_on='idcargoabonoSK',how='left')
columns1=["count","mes","nbrcargoabono"]
transac_2=merge_fact_cargoabono[columns1]

#filtramos los registros que solo sean de tipo Abono
transac_2 = transac_2[(transac_2.nbrcargoabono == 'Abono')]
transac_2
```

```python
#Transacciones mensualmente de tipo abono
transac_2=transac_2[["mes",'count']].groupby("mes").count()
transac_2=transac_2.sort_values(by=["mes"], ascending = True).reset_index()
transac_2
```

```python
#Hacer merge para traer el nombre del mes
resul_transac_2=pd.merge(transac_2,df_tiempo_mesunico, on='mes',how='left')
#resul_transac_mensual2 = resul_transac_mensual[resul_transac_mensual['_merge'] == 'left']
resul_transac_2
```

```python
#visualizaciòn

import matplotlib.pyplot as plt

# Datos
x = resul_transac_2['mes_nombre']
y = resul_transac_2['count']

# Crear el diagrama de barras
plt.figure(figsize=(20,10))
plt.bar(x, y, color='green')

# Añadir etiquetas y título
plt.xlabel('Meses')
plt.ylabel('Ctd Trx')
plt.title('Tendencia de la ctd trx de tipo Abono en los meses del año')

# Mostrar el gráfico
plt.show()
```

### 3.Cuál es la sucursal con mayor cantidad de  transacciones ?

```python
#Agrupamos por sucursal 
transac_3=df_fact[["DIM_SUCURSAL_idsucursalSK","count"]].groupby("DIM_SUCURSAL_idsucursalSK").count()
transac_3=transac_3.sort_values(by=["count"], ascending = False).reset_index()
transac_3
```

```python
#creamos el dataframe de las trx
merge_df_sucursal=pd.merge(transac_3,df_dimsucursal,left_on='DIM_SUCURSAL_idsucursalSK',right_on='idsucursalSK',how='left')
columns2=["nbrsucursal","count"]
transac_3=merge_df_sucursal[columns2]
transac_3
```

```python
#obtenemos la sucursal con mayores cantidades de trx en el año
max_sucursal = transac_3['count'].max()
resul_transac_3=transac_3.loc[transac_3['count']==max_sucursal]
resul_transac_3

```

```python
#visualizaciòn

import matplotlib.pyplot as plt

# Datos
x = transac_3['nbrsucursal']
y = transac_3['count']

# Crear el gráfico de dispersión
plt.figure(figsize=(35,10))
plt.scatter(x, y)

# Añadir etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')

# Mostrar el gráfico
plt.show()
```

```python
#visualizaciòn

import matplotlib.pyplot as plt

# Datos
categorias = transac_3['nbrsucursal']
valores = transac_3['count']

#colores
plt.figure(figsize=(20,10))
colors=['gold','yellowgreeen','lightcoral','lightskyblue']


# Crear el diagrama de barras
#plt.bar(categorias, valores, color='green')
plt.pie(valores,labels=categorias,  autopct='%.2f%%', pctdistance=0.8)
plt.axis("equal")
plt.show()
```

### 4. ¿Cuánto es la ctd de transaccines sin éxito por tipo de transacciòn en el año ?

```python
#creamos el dataframe de las trx
merge_fact_tiptran=pd.merge(df_fact,df_tiptran,left_on='DIM_TIPOTRANSACCION_idtipotransaccionSK',right_on='idtipotransaccionSK',how='left')
columns3=["count","nbrtiptransaccion","flgtranvalida"]
transac_4=merge_fact_tiptran[columns3]

#filtramos los registros que solo sean de tipo de transaccion invalida
transac_4 = transac_4[(transac_4.flgtranvalida ==0)]
transac_4
```

```python
#Transacciones agrupados por tipotransaccion con flgtranvalida sin èxito
resul_transac_4=transac_4[["nbrtiptransaccion",'count']].groupby("nbrtiptransaccion").count()
resul_transac_4=resul_transac_4.sort_values(by=["count"], ascending = True).reset_index()
resul_transac_4
```

```python
#visualizaciòn

plt.figure(1 , figsize = (20 , 10))
sns.barplot(x =resul_transac_4['count'] , y =  resul_transac_4['nbrtiptransaccion'] , palette = "Blues_d")
```

### 5. Cuanto es el promedio de los montos que se transaccionan por canal?

```python
df_dimcanal
```

```python


merge_fact_canal=pd.merge(df_fact,df_dimcanal,left_on='DIM_CANAL_idcanalSK',right_on='idcanalSK',how='left')
columns4=["nbrcanal","monto"]
transac_5=merge_fact_canal[columns4]

transac_5
```

```python
#verificar los tipos de datos de mi dataframe
transac_5.dtypes
```

```python
 #cambiar la variable monto de object a numerico
transac_5['monto'] = pd.to_numeric(transac_5['monto'])
```

```python
#verificamos si cambio el tipo de dato de mi variable monto
transac_5.dtypes
```

```python

#Obtener el promedio del monto por canal que se transacciona al año
monto_promedio_por_canal=transac_5.groupby("nbrcanal").mean()
resul_transac_5=monto_promedio_por_canal.sort_values(by=["monto"], ascending = False).reset_index()
resul_transac_5

```

```python

#visualizaciòn
import matplotlib.pyplot as plt

y = resul_transac_5['monto']
x = resul_transac_5['nbrcanal']
colores = ["#FFD97D","#FF9B85"]
plt.pie(y, labels=x, autopct="%0.1f %%", colors=colores)
plt.axis("equal")
plt.show()
```
