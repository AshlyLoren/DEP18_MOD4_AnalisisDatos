## Crearemos la clase ConnectionMySQL

```python
# este es el fichero para conectarnos a MySQL
class ConexionMySQL():
    def __init__(self, bd_servidor,bd_nombre,bd_usuario,bd_contrasenia):
        self.bd_servidor=bd_servidor
        self.bd_nombre=bd_nombre
        self.bd_usuario=bd_usuario
        self.bd_contrasenia=bd_contrasenia

    ## Función de conexión a MySQL
    def conectarMySQL(self):
        engine = db.create_engine(f"mysql+mysqlconnector://{self.bd_usuario}:{self.bd_contrasenia}@{self.bd_servidor}:3310/{self.bd_nombre}")
        conn=engine.connect()
        return conn
```

## Cargar datos a DIM_Cliente

1.Extraeremos datos de la base de datos transaccional sobre Clientes

2. Insertamos estos valores en la DIM_Cliente del datawarehouse  

```python
dimcliente= Table('DIM_CLIENTE', metadata, autoload_with=engine)

# Create a connection to the database
conn = engine.connect()

for i,cliente in dimcliente_data.iterrows():
    # Sample data to insert
    data_to_insert = {
        "idclienteSK": cliente["idclientePK"],
        "nrodocumento": cliente["nrodocumento"],
        "nbrcli": cliente["nbrcli"],
        "email": cliente["email"],
        "nrocuenta": cliente["nrocuenta"],
    }

    try:
        # Insert data into the table
        ins = dimcliente.insert().values(data_to_insert)
        conn.execute(ins)
        print(conn)

        # Commit the transaction
        conn.commit()
        print("se envió el commit a la base  de datos")

    except Exception as e:
        # Handle exceptions
        print(f"Error: {e}")
        # Rollback the transactio

# Close the connection
conn.close()
```

## Crear la DIM_TIEMPO 

1.Creamos la dimension Tiempo

```python
import pandas as pd
import psycopg2
from datetime import datetime, timedelta

# Crear un DataFrame de fechas desde una fecha inicial hasta una fecha final
fecha_inicial = datetime(2023, 1, 1)
fecha_final = datetime(2023, 12, 31)
delta = timedelta(days=1)
fechas = []

idfecha=1
idfechas = []

while fecha_inicial <= fecha_final:
    fechas.append(fecha_inicial)
    fecha_inicial += delta
    idfechas.append(idfecha)
    idfecha=idfecha+1


# Convertir la lista de fechas en un DataFrame
df_fechas = pd.DataFrame(fechas, columns=['fecha'])



#convertir la lista de ids en un dataframe
df_ids = pd.DataFrame(idfechas, columns=['idfecha'])
df_fechas['idfechaSK'] = df_ids['idfecha']


# Extraer información de las fecha
df_fechas['año'] = df_fechas['fecha'].dt.year
df_fechas['mes'] = df_fechas['fecha'].dt.month
df_fechas['día'] = df_fechas['fecha'].dt.day
df_fechas['semana'] = df_fechas['fecha'].dt.isocalendar().week
df_fechas['mes_nombre'] = df_fechas['fecha'].dt.month_name()
df_fechas['día_nombre'] = df_fechas['fecha'].dt.day_name()

print("Dimensión de tiempo creada exitosamente.")
```

```python
dimtiempo= Table('DIM_TIEMPO', metadata, autoload_with=engine)

# Create a connection to the database
conn = engine.connect()

for i,tiempo in df_fechas.iterrows():
    # Sample data to insert
    data_to_insert = {
        "idfechaSK": tiempo["idfechaSK"],
        "fecha": tiempo["fecha"],
        "anio": tiempo["año"],
        "mes": tiempo["mes"],
        "dia": tiempo["día"],
        "semana": tiempo["semana"],
        "mes_nombre": tiempo["mes_nombre"],
        "dia_semana": tiempo["día_nombre"],
    }

    try:
        # Insert data into the table
        ins = dimtiempo.insert().values(data_to_insert)
        conn.execute(ins)
        print(conn)

        # Commit the transaction
        conn.commit()
        print("se envió el commit a la base  de datos")

    except Exception as e:
        # Handle exceptions
        print(f"Error: {e}")
        # Rollback the transactio

# Close the connection
conn.close()
```

2. Vamos a generar aleatoriamente los datos fecha para cada registro de transacción

```python
import random
from datetime import datetime, timedelta
import random

def gen_timestamp():
    # Generar un timestamp aleatorio dentro de un rango específico
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 12, 31)

    # Calcular un valor aleatorio entre start_date y end_date
    random_date = start_date + timedelta(seconds=random.randint(0, int((end_date - start_date).total_seconds())))
    random_date.date
 
    # Mostrar el timestamp aleatorio
    return random_date
```

```python
df_transac["fecha"]=df_transac["idtransaccionPK"].apply(lambda x: gen_timestamp())
df_transac["fecha"]=pd.to_datetime(df_transac["fecha"])
df_transac["fecha"]=df_transac["fecha"].dt.date
df_transac["fecha"]=pd.to_datetime(df_transac["fecha"])
df_transac['año'] = df_transac["fecha"].dt.year
df_transac["fecha"]=pd.to_datetime(df_transac["fecha"])
df_transac['mes'] = df_transac["fecha"].dt.month
df_transac["fecha"]=pd.to_datetime(df_transac["fecha"])
df_transac['día'] = df_transac["fecha"].dt.day

df_transac
```
